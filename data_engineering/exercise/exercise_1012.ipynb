```python
"""
Introduction to Broadcasting
"""

# Broadcasting : 2D + 1D arrays

x = np.ones((3,3))
print(x)

y= np.array([1,2,3])
print(y)
print(x+y)
```

```python
# Broadcast both arrays

a= np.arange(3)
print(a)
a= a[:, np.newaxis]
print(a)

b= np.arange(3)
print(b)

print(a+b)
```

```python
"""
Comparisons, Masks, and Boolean Logic
"""

x= np.array([1,2,3,4,5])
print(x)

print(x<3)
x>4
print(x>=5)
```

```python
np.random.seed(0)
x= np.random.randint(10,size=(3,4))
print(x)
print(x<6)
```

```python
"""
Working with Boolean Arrays
"""

print(x)
print(np.count_nonzero(x<6))
print(np.sum(x<6))

print(np.any(x>9))
print(np.all(x<10))
```

```python
# Masking operations

print(x>5)

print(np.min(x[x<5]))
```

```python
"""
Fancy Indexing
"""

np.random.seed(1)
x=np.random.randint(100,size=10)
print(x)

print(x[2], x[4], x[6])

ind=[2, 4, 6]
print(x[ind])
```

```python
# For n-D arrays

np.random.seed(2)
x = np.random.randint(10, size=(3,5))
print(x)

row = [0,1,2]
col = [1,3,2]
x[row,col]
```

```python
# For n-D arrays with broadcasting

print(x)
row= np.array([0,1,2])
col= np.array([2,1,4])
x[row,col]
print(x[row[:,np.newaxis],col])
```

```python
"""
Combinded Indexing
"""

x= np.arange(1,13).reshape(4,3)
print(x)

print(x[2,[2,0,1]])
```

```python
"""
Example: Selecting Random Points
"""
mean = [0,0]
cov= [[1,2],
      [2,5]]

rand = np.random.RandomState(1)
x= rand.multivariate_normal(mean,cov,100)
print(x)
print(x.shape)
```

```python
plt.scatter(x[:,0],x[:,1]);
```

```python
ind = np.random.choice(x.shape[0], 20, replace=False)
print(ind)
```

```python
selected= x[ind]
print(selected)
print(selected.shape)
```

```python
plt.scatter(x[:,0],x[:,1],alpha=0.3)
plt.scatter(selected[:,0], selected[:,1],
           facecolor='red', s=50)
```
